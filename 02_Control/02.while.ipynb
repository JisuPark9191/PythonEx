```python
num = 1
while num <= 10:
    print(f'나무를 {num}번 찍는다')
    num = num + 1

print('end~')
```

```python
# 무한루프

while True:
    money = int(input('현재 금액은?'))

    if money == -1:
        print('탈출~')
        break

    if money >= 10000:
        print('케익 사먹을 수 있다 ^^')
    elif money >= 5000:
        print('담배 사 필 수 있다 ^^')
    elif money >= 3000:
        print('콜라 마실 수 있다 ^^')
    elif money >= 1200:
        print('빵 사먹을 수 있다 ^^')
    else:
        print('집에서 밥 먹어!')

print('end...')
```
